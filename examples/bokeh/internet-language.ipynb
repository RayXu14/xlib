```python
import pandas as pd

from bokeh.plotting import figure, show, output_notebook, output_file
from bokeh.tile_providers import get_provider, Vendors
from bokeh.models import ColumnDataSource, LabelSet, Select, CDSView, IndexFilter
from bokeh.layouts import column, row
from bokeh.transform import dodge
from bokeh.palettes import Spectral10, viridis, Paired10, Viridis11, cividis, Spectral11, Cividis11, Category10

OUTPUT_PATH = 'world-internet-language.html'

output_notebook()
output_file(OUTPUT_PATH)
```

```python
top_languages = ['English', 'Chinese', 'Spanish', 'Arabic', 'Portuguese', 
                 'Indonesia', 'French', 'Japanese', 'Russian', 'German']
internet_population = [1105, 863, 344, 226, 171, 169, 144, 118, 109, 92]
colors = Category10[10]
top_languages.reverse()
internet_population.reverse()
colors.reverse()
internet_population_str = [str(e) for e in internet_population]

source = ColumnDataSource(data=dict(top_lang=top_languages,
                                    net_pop=internet_population,
                                    net_pop_str=internet_population_str,
                                    color=colors))
```

```python
total_rank = figure(y_range=top_languages, x_range=(0, 1300),
                    title='Top 10 Language Used on the Internet',
                    tools='', toolbar_location=None,
                    height=500, width=1000) #ã€€y_axis_location="right"
total_rank.xaxis.axis_label = 'Millions of Users'
# total_rank.yaxis.axis_label = 'Languages'

total_rank.hbar(y='top_lang', right='net_pop', source=source,
                color='dodgerblue', height=0.5, alpha=.8) # color='crimson', 
label_score = LabelSet(x='net_pop', y='top_lang', text='net_pop_str',
                       source=source, x_offset=+10, y_offset=-15,
                       text_font_size='18pt', text_font_style='bold')
total_rank.add_layout(label_score)

total_rank.title.text_font_size = '24pt'
total_rank.axis.axis_label_text_font_size = '22pt'
total_rank.axis.major_label_text_font_size = '18pt'
# total_rank.xaxis.axis_label_text_font_size = '12pt'
# total_rank.xaxis.major_label_text_font_size = '10pt'
total_rank.title.text_font_style = 'bold'
total_rank.axis.axis_label_text_font_style = 'bold'
total_rank.axis.major_label_text_font_style = 'bold'

# total_rank.grid.grid_line_color = None
total_rank.ygrid.grid_line_color = None
total_rank.axis.axis_line_color = None
total_rank.outline_line_color = None
total_rank.axis.minor_tick_line_color = None
total_rank.axis.major_tick_line_color = None
show(total_rank)
```
